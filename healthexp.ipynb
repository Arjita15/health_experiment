```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
def load_dataset(file_path):
 return pd.read_csv(file_path)
df = load_dataset("./Health_heart_experimental.csv")
df.head(5)
```

```python
#line plot
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
nump = [5000, 10000, 18000, 27000, 40000, 55000, 72000]
plt.plot(years, nump, color='green', marker='+', linestyle='dotted')
plt.title("Health Experiment Data By years")
plt.ylabel("No.of patient flow")
plt.xlabel("Years")
plt.show()
```

```python
#line plot
sns.lineplot(x='BMI',y='sex',data=df)
```

```python
#bar plot
sysbp= ["80", "95", "110", "125", "140" ,"165","180"]
age = [13,39,90,65,78,52,26]
plt.bar(range(len(sysbp)), age)
plt.title("Health Experimental")     
plt.ylabel("#Age of patient")   
plt.xlabel("sysBP level")
plt.xticks(range(len(sysbp)), sysbp)
```

```python
#bar plot
sns.barplot(x='sex',y='age',hue='indication',data=df)
```

```python
#pie plot
df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
#box plot
sns.boxplot(df['age'])
```

```python
#box plot
sns.boxplot(x='sex',y='age',hue='indication',data=df)
```

```python
#histogram
plt.hist(df['BMI'])
```

```python
#histogram
sns.kdeplot(df[df['indication'] == 0]['BMI'], label='0')
sns.kdeplot(df[df['indication'] == 1]['BMI'], label='1')
plt.title('BMI Distribution by indication')
plt.xlabel('BMI')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
#histogram
sns.displot(df['BMI'], kind="hist", kde=True)
```

```python
#scatter plot
sns.scatterplot(x='weightKg', y='heightCm',hue='sex', style='indication',data=df)
plt.show()
```
